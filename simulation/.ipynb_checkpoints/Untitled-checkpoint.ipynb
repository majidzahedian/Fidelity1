```python
P = 0.04 # mW
la1 = 20823.89228823*P  # cps
la2 = 160507.64200232*P # cps
ga1 = 3872.39255739*P*P    # Hz
ga2 = 24142.64150491 *P*P  # Hz
TR = 11e-3  # Read out time is second
print(ga2)
a=np.array([1,2,3])
a= np.append(a,5)
a
```

```python
##########################################################################################################
def integrand_i1(t,n):
    a_odd  =  ga1*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*t/(TR-t))*iv(1,2*sqrt(ga1*ga2*t*(TR-t))) 
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even+a_odd)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_i1(n):
    return quad(integrand_i1, 0, TR, args=n)[0] + exp(-ga1*TR)*poisson.pmf(n, la1*TR)
#----------------------------------
def integrand_i2(t,n):
    a_odd  =  ga2*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*(TR-t)/t)*iv(1,2*sqrt(ga1*ga2*t*(TR-t)))
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even+a_odd)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_i2(n):
    return quad(integrand_i2, 0, TR, args=n)[0] + exp(-ga2*TR)*poisson.pmf(n, la2*TR)

###############################################################################################################
def integrand_f1(t,n):
    a_odd = ga2*iv(0,2*sqrt(ga1*ga2*t*(TR-t)))
    a_even = sqrt(ga1*ga2*t/(TR-t))*iv(1,2*sqrt(ga1*ga2*t*(TR-t)))
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even*p1+a_odd*p2)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_f1(n):
    return  quad(integrand_f1, 0, TR, args=n)[0] + exp(-ga1*TR)*p1*poisson.pmf(n, la1*TR)
#----------------------------------
def integrand_f2(t,n):
    a_odd = ga1*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*(TR-t)/t)*iv(1,2*sqrt(ga1*ga2*t*(TR-t))) 
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even*p2+a_odd*p1)*poisson.pmf(n, la1*t+la2*(TR-t))
    
def P_n_Cond_f2(n):
    return  quad(integrand_f2, 0, TR, args=n)[0] + exp(-ga2*TR)*p2*poisson.pmf(n, la2*TR)
###############################################################################################################

def P_n_1(n):
    return poisson.pmf(n, la1*TR)
#----------------------------------
def P_n_2(n):
    return poisson.pmf(n, la2*TR)
##############################################################################################################
## finding intersection of two curves as threshold for fidelity
def n_th(arr1,arr2):
    i = argmax(arr1)
    j = argmax(arr2)
    return i+argmin(np.abs(arr1[i:j]-arr2[i:j]))
###############################################################################################################

def Fidelity_i():
    prob_n_Cond_i1 = np.array([])
    prob_n_Cond_i2 = np.array([])
    N = np.arange(poisson.ppf(0.0001, la1*TR),poisson.ppf(0.9999, la2*TR))
    for n in N:
        prob_n_Cond_i1 = np.append(prob_n_Cond_i1, P_n_Cond_i1(n))
        prob_n_Cond_i2 = np.append(prob_n_Cond_i2, P_n_Cond_i2(n))
        #prob_n_Cond_i1.append(P_n_Cond_i1(n))
        #prob_n_Cond_i2.append(P_n_Cond_i2(n))

    prob_n_Cond_i1 = prob_n_Cond_i1/np.trapz(prob_n_Cond_i1)
    prob_n_Cond_i2 = prob_n_Cond_i2/np.trapz(prob_n_Cond_i2)
    n_th_i = n_th(prob_n_Cond_i1,prob_n_Cond_i2)
    F_i = 1 - np.trapz(prob_n_Cond_i2[0:n_th_i+1]) - np.trapz(prob_n_Cond_i1[n_th_i:])
    return F_i
###########################################################################################################
def Fidelity_f():
    prob_n_Cond_f1 = []
    prob_n_Cond_f2 = []
    N = np.arange(poisson.ppf(0.0001, la1*TR),poisson.ppf(0.9999, la2*TR))
    for n in N:
        prob_n_Cond_f1.append(P_n_Cond_f1(n))
        prob_n_Cond_f2.append(P_n_Cond_f2(n))

    prob_n_Cond_f1 = np.array(prob_n_Cond_f1)
    prob_n_Cond_f2 = np.array(prob_n_Cond_f2)
    prob_n_Cond_f1 = prob_n_Cond_f1/np.trapz(prob_n_Cond_f1)
    prob_n_Cond_f2 = prob_n_Cond_f2/np.trapz(prob_n_Cond_f2)
    n_th_f = n_th(prob_n_Cond_f1,prob_n_Cond_f2)
    F_f = 1 - np.trapz(prob_n_Cond_f2[0:n_th_f+1]) - np.trapz(prob_n_Cond_f1[n_th_f:])
    return F_f
##################################################################################################
```

```python
def integrand_Pn_even(t,n,la1,la2,g1,g2,TR):
    return exp((ga1-ga2)*t-ga1*TR)*sqrt(ga1*ga2*t/(TR-t))*iv(1,2*sqrt(ga1*ga2*t*(TR-t)))*poisson.pmf(n, la2*t+la1*(TR-t))
    
def Pn_even(n,la1,la2,g1,g2,TR): # probability to emit n photons for even jumps FROM LEVEL 2 TO LEVEL 1
    return quad(integrand_Pn_even, 0, TR, args=[n,la1,la2,g1,g2,TR])[0] + exp(-ga2*TR)*poisson.pmf(n, la2*TR)

##########################################################################################################
def integrand_Pn_odd(t,n,la1,la2,g1,g2,TR):
    return exp((ga1-ga2)*t-ga1*TR)*iv(0,2*sqrt(ga1*ga2*t*(TR-t)))*poisson.pmf(n, la2*t+la1*(TR-t))
    
def Pn_odd(n,la1,la2,g1,g2,TR): # probability to emit n photons for odd jumps FROM LEVEL 2 TO LEVEL 1
    return quad(integrand_Pn_odd, 0, TR, args=[n,la1,la2,g1,g2,TR])[0]

##########################################################################################################
def integrand_i1(t,n):
    a_odd  =  ga1*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*t/(TR-t))*iv(1,2*sqrt(ga1*ga2*t*(TR-t))) 
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even+a_odd)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_i1(n):
    return quad(integrand_i1, 0, TR, args=n)[0] + exp(-ga1*TR)*poisson.pmf(n, la1*TR)
#----------------------------------
def integrand_i2(t,n):
    a_odd  =  ga2*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*(TR-t)/t)*iv(1,2*sqrt(ga1*ga2*t*(TR-t)))
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even+a_odd)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_i2(n):
    return quad(integrand_i2, 0, TR, args=n)[0] + exp(-ga2*TR)*poisson.pmf(n, la2*TR)

###############################################################################################################
def integrand_f1(t,n):
    a_odd = ga2*iv(0,2*sqrt(ga1*ga2*t*(TR-t)))
    a_even = sqrt(ga1*ga2*t/(TR-t))*iv(1,2*sqrt(ga1*ga2*t*(TR-t)))
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even*p1+a_odd*p2)*poisson.pmf(n, la1*t+la2*(TR-t))

def P_n_Cond_f1(n):
    return  quad(integrand_f1, 0, TR, args=n)[0] + exp(-ga1*TR)*p1*poisson.pmf(n, la1*TR)
#----------------------------------
def integrand_f2(t,n):
    a_odd = ga1*iv(0,2*sqrt(ga1*ga2*t*(TR-t))) 
    a_even = sqrt(ga1*ga2*(TR-t)/t)*iv(1,2*sqrt(ga1*ga2*t*(TR-t))) 
    return exp((ga2-ga1)*t)*exp(-ga2*TR)*(a_even*p2+a_odd*p1)*poisson.pmf(n, la1*t+la2*(TR-t))
    
def P_n_Cond_f2(n):
    return  quad(integrand_f2, 0, TR, args=n)[0] + exp(-ga2*TR)*p2*poisson.pmf(n, la2*TR)
###############################################################################################################

def P_n_1(n):
    return poisson.pmf(n, la1*TR)
#----------------------------------
def P_n_2(n):
    return poisson.pmf(n, la2*TR)
##############################################################################################################
## finding intersection of two curves as threshold for fidelity
def n_th(arr1,arr2):
    i = argmax(arr1)
    j = argmax(arr2)
    return i+argmin(np.abs(arr1[i:j]-arr2[i:j]))
###############################################################################################################

def Fidelity_i(la1,la2,g1,g2,TR):
    prob_n_Cond_i1 = np.array([])
    prob_n_Cond_i2 = np.array([])
    N = np.arange(poisson.ppf(0.0001, la1*TR),poisson.ppf(0.9999, la2*TR))
    for n in N:
        prob_n_Cond_i1 = np.append(prob_n_Cond_i1, P_n_Cond_i1(n))
        prob_n_Cond_i2 = np.append(prob_n_Cond_i2, P_n_Cond_i2(n))
        #prob_n_Cond_i1.append(P_n_Cond_i1(n))
        #prob_n_Cond_i2.append(P_n_Cond_i2(n))

    prob_n_Cond_i1 = prob_n_Cond_i1/np.trapz(prob_n_Cond_i1)
    prob_n_Cond_i2 = prob_n_Cond_i2/np.trapz(prob_n_Cond_i2)
    n_th_i = n_th(prob_n_Cond_i1,prob_n_Cond_i2)
    F_i = 1 - np.trapz(prob_n_Cond_i2[0:n_th_i+1]) - np.trapz(prob_n_Cond_i1[n_th_i:])
    return F_i
###########################################################################################################
def Fidelity_f():
    prob_n_Cond_f1 = []
    prob_n_Cond_f2 = []
    N = np.arange(poisson.ppf(0.0001, la1*TR),poisson.ppf(0.9999, la2*TR))
    for n in N:
        prob_n_Cond_f1.append(P_n_Cond_f1(n))
        prob_n_Cond_f2.append(P_n_Cond_f2(n))

    prob_n_Cond_f1 = np.array(prob_n_Cond_f1)
    prob_n_Cond_f2 = np.array(prob_n_Cond_f2)
    prob_n_Cond_f1 = prob_n_Cond_f1/np.trapz(prob_n_Cond_f1)
    prob_n_Cond_f2 = prob_n_Cond_f2/np.trapz(prob_n_Cond_f2)
    n_th_f = n_th(prob_n_Cond_f1,prob_n_Cond_f2)
    F_f = 1 - np.trapz(prob_n_Cond_f2[0:n_th_f+1]) - np.trapz(prob_n_Cond_f1[n_th_f:])
    return F_f
##################################################################################################
```

```python
P = 0.01888 # mW
la1 = 20823.89228823*P  # cps
la2 = 160507.64200232*P # cps
ga1 = 3872.39255739*P*P    # Hz
ga2 = 24142.64150491 *P*P  # Hz
TR = 5e-3  # Read out time is second
p2 = ga1/(ga1+ga2)
#p2 = 0.75
p1 = 1-p2
print(p1,p2)
prob_n_Cond_i1 = []
prob_n_Cond_i2 = []
prob_n_Cond_f1 = []
prob_n_Cond_f2 = []
prob_n_1 = []
prob_n_2 = []
N = np.arange(poisson.ppf(0.0001, la1*TR),poisson.ppf(0.9999, la2*TR))
for n in tqdm.tqdm_notebook(N):
    prob_n_Cond_i1.append(P_n_Cond_i1(n))
    prob_n_Cond_i2.append(P_n_Cond_i2(n))
    prob_n_Cond_f1.append(P_n_Cond_f1(n))
    prob_n_Cond_f2.append(P_n_Cond_f2(n))
    prob_n_1.append(P_n_1(n))
    prob_n_2.append(P_n_2(n))

prob_n_1 = np.array(prob_n_1)
prob_n_2 = np.array(prob_n_2)
prob_n_Cond_i1 = np.array(prob_n_Cond_i1)
prob_n_Cond_i2 = np.array(prob_n_Cond_i2)
prob_n_Cond_f1 = np.array(prob_n_Cond_f1)
prob_n_Cond_f2 = np.array(prob_n_Cond_f2)

prob_n_Cond_f1 = prob_n_Cond_f1/np.trapz(prob_n_Cond_f1)
prob_n_Cond_f2 = prob_n_Cond_f2/np.trapz(prob_n_Cond_f2)
prob_n_Cond_i1 = prob_n_Cond_i1/np.trapz(prob_n_Cond_i1)
prob_n_Cond_i2 = prob_n_Cond_i2/np.trapz(prob_n_Cond_i2)
```

# Fidelity for initial states

```python
F_2d_i=[]
PP = np.linspace(0.015,0.05,10)
for P in tqdm.tqdm_notebook(PP):
    la1 = 20823.89228823*P  # cps
    la2 = 160507.64200232*P # cps
    ga1 = 3872.39255739*P*P   # Hz
    ga2 = 24142.64150491*P*P  # Hz
    p1 = ga2/(ga1+ga2)
    p2 = ga1/(ga1+ga2)
    F_1d=[]
    TTR = np.linspace(1e-3,5e-3,20)
    for TR in TTR:
        F_1d.append(Fidelity_i())
    F_2d_i.append(F_1d)
F_2d_i=np.array(F_2d_i)
```

```python
time,power=np.meshgrid(TTR, PP)
opt_F = np.max(F_2d_i)
print("Max Fidelity = {}".format(opt_F))
result=np.where(F_2d_i == np.amax(F_2d_i))
listOfCordinates = list(zip(result[0], result[1]))
for cord in listOfCordinates:
    if F_2d_i[cord[0]][cord[1]]==np.max(F_2d_i):
        opt_p = power[cord[0]][cord[1]]
        opt_t = time[cord[0]][cord[1]]
        print('Optimum power = {} uW'.format(opt_p*1000))
        print('Optimum time = {} ms'.format(opt_t*1000))
    else: print('max did not calculated correctly')
   
plt.figure(dpi=100)
plt.contourf(power*1000, time*1000, F_2d_i,80)
plt.title('Fidelity (orange, initial=final)', fontsize =18)
plt.plot(opt_p*1000, opt_t*1000, '*', ms=10, color='red', label='Optimum')
plt.xlabel('Power (uW)', fontsize=16)
plt.ylabel('Readout Time (ms)', fontsize =16)
plt.colorbar()
plt.legend()
#plt.savefig("F_initial_state.png",dpi=150)

plt.figure()
plt.plot(N, prob_n_Cond_f1, '-o', ms=3, label='final NV0')
plt.plot(N, prob_n_Cond_f2, '-o', ms=3, label='final NV-')
plt.vlines(6,ymin=0, ymax=max(prob_n_Cond_f1), color='black',label='Threshold')
plt.legend(fontsize=12)
plt.xlabel('Photon Number', fontsize=16)
plt.ylabel('Histogram', fontsize=16)
plt.title('P= {} uW'.format(1e-3*round(opt_p*1e6))+', TR= {} ms'.format(opt_t*1e3)
          +', F= {} %'.format(round(opt_F*1e4)*1e-2), fontsize=16)
```
